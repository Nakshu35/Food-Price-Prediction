```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('food_prices_ind.csv')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df=df.drop(0).reset_index(drop=True)
df[['price']]=df[['price']].astype(float)
df[['usdprice']]=df[['usdprice']].astype(float)
```

```python
df.head()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.tail()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df['YEAR'] = df['date'].str.split('-').str[0]
df['MONTH'] = df['date'].str.split('-').str[1]
df['DAY'] = df['date'].str.split('-').str[2]
```

```python
df.head()
```

```python
df[['YEAR']]=df[['YEAR']].astype(int)
df[['MONTH']]=df[['MONTH']].astype(int)
df[['DAY']]=df[['DAY']].astype(int)
```

```python
df.drop('date',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.columns.str.strip()
```

```python
df.head()
```

```python
df.drop('latitude',axis=1,inplace=True)
df.drop('longitude',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['pricetype'].value_counts()
```

```python
df['pricetype'] = df['pricetype'].apply(lambda x: 0 if x == 'Retail' else 1)
```

```python
df.head()
```

```python
df.drop('priceflag',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['category'].value_counts()
```

```python
df['category'] = df['category'].map({'oil and fats':0,'cereals and tubers':1,'miscellaneous food':2,'pulses and nuts':3,'vegetables and fruits':4,'milk and dairy':5})
```

```python
df.head()
```

```python
df['commodity'].value_counts()
```

```python
unique_values = df['commodity'].unique()
value_mapping = {val: idx for idx, val in enumerate(unique_values)}
df['commodity']=df['commodity'].map(value_mapping)
```

```python
df.head()
```

```python
df['unit'].value_counts()
```

```python
df['unit'] = df['unit'].map({'KG':0,'L':1,'100 KG':2})
```

```python
df['currency'].value_counts()
```

```python
df.drop('currency',axis=1,inplace=True)
```

```python
df['admin1'].value_counts()
```

```python
unique_values = df['admin1'].unique()
value_mapping = {val: idx for idx, val in enumerate(unique_values)}
df['admin1']=df['admin1'].map(value_mapping)
```

```python
df['admin2'].value_counts()
```

```python
unique_values = df['admin2'].unique()
value_mapping = {val: idx for idx, val in enumerate(unique_values)}
df['admin2']=df['admin2'].map(value_mapping)
```

```python
df['market'].value_counts()
```

```python
unique_values = df['market'].unique()
value_mapping = {val: idx for idx, val in enumerate(unique_values)}
df['market']=df['market'].map(value_mapping)
```

```python

df.head()
```

```python
df[['admin1','admin2','market','category','commodity','unit','pricetype']] = df[['admin1','admin2','market','category','commodity','unit','pricetype']].astype(int)
```

```python
df.drop('usdprice',axis=1,inplace=True)
```

```python
df.to_csv('food_prices_ind_cleaned.csv',index=False)
```

```python
## model TRAINING
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import ElasticNet
from sklearn.neighbors import KNeighborsRegressor
```

```python
features = df.columns
```

```python
features=features.drop('price')
```

```python
target = 'price'
```

```python
X = df[features]
Y = df[target]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
Model1 = ElasticNet()
Model1.fit(X_train,Y_train)
```

```python
Model2 = KNeighborsRegressor()
Model2.fit(X_train,Y_train)
```

```python
features
```

```python
y_pred_model1 = Model1.predict(X_test)
y_pred_model2 = Model2.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print(mean_absolute_error(Y_test, y_pred_model1))
print(mean_absolute_error(Y_test,y_pred_model2))
```

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
model3 = Lasso()
model4 = Ridge()
```

```python
model3.fit(X_train,Y_train)
```

```python
model4.fit(X_train,Y_train)
```

```python
y_pred_model3 = model3.predict(X_test)
y_pred_model4 = model4.predict(X_test)
print(mean_absolute_error(Y_test, y_pred_model3))
print(mean_absolute_error(Y_test, y_pred_model4))
```
